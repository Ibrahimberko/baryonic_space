## What is an SDR?
**Software Defined Radio (SDR)** is a radio communication system where components that have been traditionally implemented in hardware (e.g.  mixers, filters, amplifiers, modulators/demodulators, detectors, etc.) are implemented in **software** instead running on a personal computer or embedded system [wikipedia](https://en.wikipedia.org/wiki/Software-defined_radio). Computation platform can be anything from general purpose CPUs to FPGAs, from GPUs to DSP chips. 

With SDR one can access some part of electromagnetic spectrum, **monitor**, **capture**, **demodulate** it. Width of the RF signal depends on the capability of RF front-end. Another important factor is resolution. 

SDR has been an important tool in education, industry for years, providing us over-the-air signals easily so that we can make sure our DSP algorithms behave the same as they do in the simulation. 

In this lab we will learn about our SDR hardware, that is **RTL-SDR** and look into SDR tools giving us access to electromagnetic spectrum. 

## RTL-SDR
In previous labs we have only used our PC with sound-card to implement some basic DSP functionalities. With this lab and forward we will use and RF front-end, **RTL-SDR**, to receive real signals around us. This device allows transferring the raw I/Q samples to us. The highest sampling rate they can provide is 3.2MSPS but we will often use lower sampling rates to be on the safe side.

Let's have a look at some of the binaries 
* **rtl_test**: this will test the device and its function.
* **rtl_sdr**: enables saving raw samples into a file
* **rtl_tcp**: networked interface that enables loading data, changing frequency, gain etc.
* **rtl_fm**: receive commercial wide-band FM signals
* **rtl_adsb**: receive ADS-B signals emitted from aircrafts

We will use **rtl_test** and **rtl_sdr** tools in this week. 

### Testing your RTL-SDR
Use `rtl_test` tool to test your device. You can also get a list of gain you can apply. 

### Capturing IQ data
You can capture IQ data using `rtl_sdr` application. If you are outside or near to window you will get better result in demodulation. If you opt to capture your own signal make sure you have applied gain suitable to receive correctly. Again you can use rtl_test to list available gains. You can set gain by using -g switch for rtl_sdr. 

Set SDR to your favorite station's frequency and capture 10 seconds of IQ data. For example to capture 10 seconds of IQ data sampled at 1.024MSPS from TRT FM radio station use an **example command** like below. 
```
rtl_sdr -f 88000000 -s 1024000 -n 10240000 wbfm88_10s.dat
``` 
This will put samples into a file called `wbfm88_10s.dat` in your current folder. 

## SDR#
We will use one of the most common general purpose SDR applications called **SDR#** in this lab. There are many general purpose SDR applications, and they do the same things more or less. We will see SDR# in action. So let's open it and do some FM-listening. This [guide](https://www.rtl-sdr.com/sdrsharp-users-guide/) has a lot of information for SDR#. 

# Task-1
## Q1: Use `rtl_test` to see if your device exists and works actually. (10pts)

## A1: 
Use the code cell below

## Q2: Use SDR# to demodulate WBFM in real-time (10pts)
This time use SDR# application to listen to **TRT Nagme** radio station. Put a **screenshot** below showing your SDR# in action.  

## A2: 
Use the code cell below

# Task-2
## Q1: Use RTL-SDR within GNU Radio (20pts)
To access RTL-SDR from GNU Radio, we must install SoapySDR RTL-SDR support file. Follow this [document](https://wiki.gnuradio.org/index.php?title=CondaInstall#Installing_related_software) to install the necessary module. You can simply install it with command: `conda install soapysdr-module-rtlsdr`

You can check whether Soapy finds your RTL-SDR by running `SoapySDRUtil --find`. Below is my output when I ran this command
