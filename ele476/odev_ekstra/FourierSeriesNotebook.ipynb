# Fourier Series Notebook

In this notebook, we approximate **periodic waveforms** using their **Fourier Series** expansions:
- Square Wave
- Triangular Wave
- Sawtooth Wave


## Fourier Series Coefficients Explanation

For a periodic waveform with period \( T = 1 / f_0 \), the **Fourier Series** is:

$$ x(t) = \sum_{k=1}^{\infty} A_k \sin(2\pi k f_0 t) $$

Here are the harmonic coefficients \($ A_k$ \) for each waveform:

### Square Wave
Only **odd** harmonics (1st, 3rd, 5th...) are present.

$A_k = \frac{4}{\pi k}, \quad k = 1, 3, 5, ...$

### Triangular Wave
Only **odd** harmonics, with **alternating signs** and decaying as \($ 1/k^2$ \):

$A_k = \frac{8}{\pi^2 k^2} (-1)^{(k-1)/2}, \quad k = 1, 3, 5, ...$


### Sawtooth Wave
All harmonics are present, alternating sign, decaying as \($ 1/k$ \):
$A_k = \frac{1}{\pi k} (-1)^{k+1}, \quad k = 1, 2, 3, 4, ...$



```python
import numpy as np
import matplotlib.pyplot as plt

t = np.linspace(0, 1, 1000)
F0 = 5
omega0 = 2 * np.pi * F0
N = 5
```

```python
def fourier_square(t, F0, N):
    result = np.zeros_like(t)
    for k in range(1, 2*N, 2):  # odd harmonics
        result += (1 / k) * np.sin(k * 2 * np.pi * F0 * t)
    return result * (4 / np.pi)

square_wave = fourier_square(t, F0, N)

plt.figure(figsize=(6, 3))
plt.plot(t, square_wave)
plt.title("Square Wave (Fourier Series)")
plt.grid(True)
plt.show()
```

```python
def fourier_triangle(t, F0, N):
    result = np.zeros_like(t)
    for k in range(1, 2*N, 2):  # odd harmonics
        result += ((-1)**((k - 1)//2)) * (1 / k**2) * np.sin(k * 2 * np.pi * F0 * t)
    return result * (8 / np.pi**2)

triangle_wave = fourier_triangle(t, F0, N)

plt.figure(figsize=(6, 3))
plt.plot(t, triangle_wave)
plt.title("Triangular Wave (Fourier Series)")
plt.grid(True)
plt.show()
```

```python
def fourier_sawtooth(t, F0, N):
    result = np.zeros_like(t)
    for k in range(1, N + 1):
        result += ((-1)**(k + 1)) * (1 / k) * np.sin(k * 2 * np.pi * F0 * t)
    return result * (2 / np.pi)

sawtooth_wave = fourier_sawtooth(t, F0, N)

plt.figure(figsize=(6, 3))
plt.plot(t, sawtooth_wave)
plt.title("Sawtooth Wave (Fourier Series)")
plt.grid(True)
plt.show()
```
