# Tune Frequency - Two Instruments!

In this module, we will analyze the tune signal produced by two instruments that are **not perfectly in tune** with each other. 

# What's a tuning tone? 

<img src="udvekanun.jpeg">
Imagine two idealized musical instruments producing pure single-pitch tones, both at a standard frequency of 440 Hz (commonly referred to as middle A). While real instruments generate additional overtones (harmonics) at integer multiples of the fundamental frequency, we will ignore those for this example. If one instrument plays slightly sharp by 5 Hz and the other slightly flat by 5 Hz, this situation can be represented mathematically as a superposition of signals.


$x(t) = \cos(2\pi 435 t) + \cos(2\pi 445 t)$

# Task-1 
## Q1: Generating the signal using Python (10 pts)

Generate half a second long signal described above, sampled at 8,000 Hz. Plot both the **half a second length waveform** and the **single-sided magnitude** of spectrum by using Python. 

## A1: 

```python
import numpy as np
import matplotlib.pyplot as plt

fs = 8000.0
f1 = 435.0
f2 = 445.0
T = 0.5  

t = np.arange(0, T, 1/fs)
x = np.cos(2*np.pi*f1*t) + np.cos(2*np.pi*f2*t)

plt.figure(figsize=(12, 4))
plt.plot(t, x)
plt.title("Time Domain Signal")
plt.xlabel("Time (s)")
plt.ylabel("Amplitude")
plt.grid(True)
plt.show()

X = np.fft.fft(x)
N = len(X)
freqs = np.fft.fftfreq(N, d=1/fs)
positive_freqs = freqs[:N//2]
magnitude = 2*np.abs(X[:N//2]) / N

plt.figure(figsize=(12, 4))
plt.plot(positive_freqs, magnitude)
plt.title("Single-Side Magnitude Spectrum")
plt.xlabel("Frequency (Hz)")
plt.ylabel("Magnitude")
plt.grid(True)
plt.show()
```

## A1: 

The superposition of 435 Hz and 445 Hz sine waves generates a 5 Hz amplitude modulation envelope on a 440 Hz carrier frequency. This beating phenomenon comes from the interference of two proximate frequencies, visible in the time domain as periodic amplitude variations. The resulting waveform appears as an AM-modulated wave, where the beat frequency — defined as the absolute difference between the two frequencies — is 10 Hz.
