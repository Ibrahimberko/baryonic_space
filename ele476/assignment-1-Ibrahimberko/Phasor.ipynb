# Phasor
When sinusoidal signals of like frequency $ω_0$ are **added together (superimposed)**, the result is a **single sinusoidal** signal having composite amplitude A and phase ϕ. In mathematical terms, given N sinusoidal functions (signals) each of identical frequency $ω_0$, you can show that the sum

$\sum_{k=1}^{N} A_k cos(w_0 t +  ϕ_k) = A cos(w_0 t +  ϕ)$	

## Example
Given $x(t) = 4 cos(w_0 t + \pi/4) + 15 sin(w_0 t)$, <br>
Let's find the composite signal $A cos(w_0 t + ϕ)$

We can follow these steps to get A and ϕ:

1. Rewrite the sine term as a cosine, so all terms in the sum fit the phasor addition formula mathematical form.

$15 sin(w_0 t) = 15 cos(w_0 t - \pi/2)$

2. Identify the magnitude and angles $A_k e^{jϕ_k} $, k = 1, 2, two terms in the problem statement:

$A_1 e^{jϕ_1} = 4 e^{j\pi/4}$ and $A_2 e^{jϕ_2} = 15 e^{-j\pi/2}$

3. Add the two complex numbers that are in polar form.

```python
%matplotlib inline
# import functions and libraries
import numpy as np, matplotlib.pyplot as plt
from matplotlib.pyplot import *
from scipy import signal
from numpy import *

A_phi=4*exp(1j*pi/4.) + 15*exp(-1j*pi/2.) 
(abs(A_phi), angle(A_phi)) # polar form (12.4959, -1.3425)
```

4. Write out the final form by putting the magnitude in front of the cosine and entering the angle inside the argument of the cosine:

$x(t) = 12.4959 cos(w_0 t - 1.3425)$
