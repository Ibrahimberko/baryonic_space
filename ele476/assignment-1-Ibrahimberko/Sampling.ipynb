# Sampling
**Nyquist Theorem** states that fs must be greater than twice the highest frequency contained in x(t) to ensure reconstruction of from its samples. For the case of a single sinusoid, this is equivalent to $F_s > 2f_0$

## Q1: What is y[n] (10 pts)

Let a continuous-time signal x(t) be defined as:

$x(t) = Re\{e^{j(2/5)\pi*t}\}$

Let y[n] be the discrete-time sequence obtained by sampling $x^2 (t)$ with the sampling period $T_s = 0.1 s$. Obtain the expression for y[n] and sketch its frequency spectrum with respect to cyclic frequency.

## A1: 

```python
import numpy as np
import matplotlib.pyplot as plt

# Parametreler
Ts = 0.1                       # Örnekleme periyodu
n = np.arange(0, 50)           # n: örnek indeksleri
t = n * Ts                     

# x(t) tanımı ve y[n] = x^2(nTs)
x = np.cos((2 * np.pi / 5) * t)
y = x**2                       # y[n] = x^2(nTs)

# Zaman domeninde y[n] çizimi
plt.figure(figsize=(8, 3))
plt.stem(n, y, basefmt=" ")
plt.title("y[n] = x²(nTs)")
plt.xlabel("n")
plt.ylabel("y[n]")
plt.grid(True)
plt.tight_layout()
plt.show()

#---------FFT-------------

Y = np.fft.fft(y)
Y_mag = np.abs(Y) / len(y)

# Cyclic frequency ekseni
freqs = np.fft.fftfreq(len(y), d=1)
freqs = np.fft.fftshift(freqs)
Y_mag = np.fft.fftshift(Y_mag)

# Spektrum
plt.figure(figsize=(8, 3))
plt.stem(freqs, Y_mag, basefmt=" ")
plt.title("Frequency Spectrum of y[n] (Cyclic Frequency)")
plt.xlabel("Cyclic Frequency (cycles/sample)")
plt.ylabel("Magnitude")
plt.grid(True)

plt.tight_layout()
plt.show()
```

As observed in the frequency spectrum, the peaks at ±0.04 cycles/sample correspond to the frequency components of the cosine term cos(4*pi/50*n).
