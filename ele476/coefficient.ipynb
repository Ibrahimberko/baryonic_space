```python
import numpy as np

# --- AYARLAR ---
derinlik = 1024  # LUT boyutu (Adres bit sayısına göre değişir)
genlik = 15000   # Maksimum genlik

# --- HESAPLAMA ---
# 0 ile 2pi arası
x = np.linspace(0, 2*np.pi, derinlik, endpoint=False)
degerler = np.round(genlik * np.sin(x)).astype(int)

# --- ÇIKTI (Sadece Sayılar) ---
# Hepsini virgülle ayrılmış tek bir blok olarak basar
print(', '.join(map(str, degerler)))
```
