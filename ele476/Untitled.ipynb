```python
import numpy as np
import matplotlib.pyplot as plt

def sinus_olustur(frekans_hz, sure_sn, ornekleme_hizi, genlik=1):
    """
    Belirli bir frekansta sinüs dalgası oluşturur.
    
    Parametreler:
    frekans_hz     : Sinyalin frekansı (Hz)
    sure_sn        : Sinyalin toplam süresi (saniye)
    ornekleme_hizi : Saniyedeki örnek sayısı (Sampling Rate - örn: 44100 veya 1000)
    genlik         : Sinyalin tepe değeri (Resimdeki 15000 gibi)
    """
    
    # Zaman vektörünü oluştur (0'dan bitiş süresine kadar)
    t = np.linspace(0, sure_sn, int(ornekleme_hizi * sure_sn), endpoint=False)
    
    # Sinüs formülü: A * sin(2 * pi * f * t)
    y = genlik * np.sin(2 * np.pi * frekans_hz * t)
    
    return t, y

# --- KULLANIM ---

# Parametreleri buradan değiştirebilirsin
istenen_frekans = 5      # 5 Hz (Saniyede 5 tur)
genlik_degeri = 15000    # Senin 'lut' değişkenindeki çarpan
sure = 1                 # 1 saniye boyunca
fs = 1000                # Örnekleme hizi (Sinyal kalitesi için)

zaman, sinyal = sinus_olustur(istenen_frekans, sure, fs, genlik_degeri)

# Çizdirme
plt.figure(figsize=(10, 4))
plt.plot(zaman, sinyal)
plt.title(f"{istenen_frekans} Hz Sinüs Dalgası")
plt.xlabel("Zaman (saniye)")
plt.ylabel("Genlik")
plt.grid(True)
plt.show()
```
